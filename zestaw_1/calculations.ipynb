```python
%%time
import numpy as np
from termcolor import colored

#Person say truth one in three attemps
p = 1/3 

#Generate set of events. Event=liars said something, which is true with probability p
nEvents = int(1E8)
nLiars = 4
events = np.random.default_rng().random((nEvents, nLiars))
events = events<p

#count number of times all liars said truth
# We get message "he sais truth", and he indeed say truth
nLastTruth = np.sum(events[:,nLiars-1]*(np.sum(events[:,:nLiars-1],axis=1)%2==1))

# We get message "he sais truth", but he lies
nLastFalse = np.sum((1-events[:,nLiars-1])*(np.sum(events[:,:nLiars-1],axis=1)%2==0))

pTruth = nLastTruth/(nLastTruth+nLastFalse)

print(colored("Probability all liars say truth:","blue"), f"{pTruth:3.4f}")
```
