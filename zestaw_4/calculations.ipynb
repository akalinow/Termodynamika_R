```python
from scipy.stats import maxwell
from scipy.constants import k as kB
from scipy.constants import proton_mass as m
import numpy as np
import matplotlib.pyplot as plt

T = 300 #K
delta_v = 201 #m/s

scale = np.sqrt(kB * T / m)
v_p = maxwell.mean(scale=scale)
v_min = v_p - delta_v / 2
v_max = v_p + delta_v / 2
r = maxwell.cdf(v_max, scale=scale) - maxwell.cdf(v_min, scale=scale)
print("Ratio = {:3.2f}".format(r))
```

```python
from scipy.stats import maxwell
from scipy.constants import k as kB
from scipy.constants import proton_mass as m
from scipy.constants import electron_volt as eV
import numpy as np

T = 1.5e7
#E = 1.8E6 #eV
E = 493E3*eV

scale = np.sqrt(kB*T/m)
v0 = np.sqrt(2*E/m)
r = 1 - maxwell.cdf(v0, scale=scale)
print("Ratio = {:3.16e}".format(r))
```
